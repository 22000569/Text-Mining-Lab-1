### Problema 1: Desarrolle una funcion la cual reciba dos parametros, una imagen y un entero llamado color, la funcion debe devolver una imagen la cual tenga activos los canales de color segun los siguientes puntos:

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
def activar_canal_color(imagen, color):
    img = cv2.imread(imagen)

    if color == 1:  # Azul
        mask = np.zeros_like(img)
        mask[:, :, 2] = img[:, :, 2]

    elif color == 2:  # Verde
        mask = np.zeros_like(img)
        mask[:, :, 1] = img[:, :, 1]

    elif color == 3:  # Rojo 
        mask = np.zeros_like(img)
        mask[:, :, 0] = img[:, :, 0]

    elif color == 10:  # Rojo y Verde
        mask = np.zeros_like(img)
        mask[:, :, 0] = img[:, :, 0]
        mask[:, :, 1] = img[:, :, 1]

    elif color == 20:  # Verde y Azul
        mask = np.zeros_like(img)
        mask[:, :, 1] = img[:, :, 1]
        mask[:, :, 2] = img[:, :, 2]

    elif color == 30:  # Azul y Rojo
        mask = np.zeros_like(img)
        mask[:, :, 0] = img[:, :, 0]
        mask[:, :, 2] = img[:, :, 2]

    else:
        raise ValueError("El valor del 'color' no es correcto.")

    imagen_final = cv2.bitwise_and(img, mask)

    return imagen_final
```

```python
imagen = "Imagen1.jpg"
Seleccionar_Color = 1
imagen_final = activar_canal_color(imagen, Seleccionar_Color)

plt.imshow(imagen_final)
plt.show()
```

### Problema 2: En el .zip del laboratorio se le compartio un conjunto de imagenes en escala de grises, estas imagenes fueron creadas utilizando una escala de grises en 3D, cree una funcion que dadas las 3 imagenes se construya la imagen original a color.

```python
def construir_imagen_color(img_red, img_green, img_blue):

    red = cv2.imread(img_red, cv2.IMREAD_GRAYSCALE)
    green = cv2.imread(img_green, cv2.IMREAD_GRAYSCALE)
    blue = cv2.imread(img_blue, cv2.IMREAD_GRAYSCALE)


    imagen_color = cv2.merge([red, green, blue])

    return imagen_color
```

### Problema 3: Cree una funcion que dada una imagen cree una escala de grises en tres dimensiones, tome en cuenta que su funcion debe crear 3 imagenes como salida. Para entregar este ejercicio debe incluir una las imagenes que haya utilizado como prueba y el resultado de las misma, no puede utilizar la imagen del Problema #2.

```python
def escala_de_grises_tres_tipos(imagen):

    img = cv2.imread(imagen)

    blue, green, red = cv2.split(img)

    img_blue_gray = cv2.cvtColor(blue, cv2.COLOR_GRAY2BGR)
    img_green_gray = cv2.cvtColor(green, cv2.COLOR_GRAY2BGR)
    img_red_gray = cv2.cvtColor(red, cv2.COLOR_GRAY2BGR)

    return img_blue_gray, img_green_gray, img_red_gray
```

### Problema 4: Cree una funcion que dada una imagen, muestre el histograma de cada canal de color y el de escala de grises (utilice un promedio aritmetico para su escala de grises, no puede usar funciones de opencv), sus histogramas deben incluir una lınea vertical la cual muestre el valor de la media de la distribucion.

```python
def Problema4_histograma(imagen):

    img = cv2.imread(imagen)

    img_gray = np.mean(img, axis=2, dtype=np.uint8) #Convertir la imagen a escala de grises 
    hist_gray, bins_gray = np.histogram(img_gray.flatten(), bins=256, range=[0, 256]) #Histograma de escala de grises

    #Histograma de cada color (rojo, verde, azul)
    hist_red, bins_red = np.histogram(img[:, :, 2].flatten(), bins=256, range=[0, 256])
    hist_green, bins_green = np.histogram(img[:, :, 1].flatten(), bins=256, range=[0, 256])
    hist_blue, bins_blue = np.histogram(img[:, :, 0].flatten(), bins=256, range=[0, 256])

    #Calcular la media de cada distribución
    mean_gray = np.mean(img_gray)
    mean_red = np.mean(img[:, :, 2])
    mean_green = np.mean(img[:, :, 1])
    mean_blue = np.mean(img[:, :, 0])

    #Crear las graficas y líneas verticales de la media
    plt.figure(figsize=(10, 6))
    plt.subplot(2, 2, 1)
    plt.plot(hist_gray, color='black')
    plt.axvline(x=mean_gray, color='blue', linestyle='--')
    plt.title('Escala de Grises')

    plt.subplot(2, 2, 2)
    plt.plot(hist_red, color='red')
    plt.axvline(x=mean_red, color='black', linestyle='--')
    plt.title('Histograma Rojo')

    plt.subplot(2, 2, 3)
    plt.plot(hist_green, color='green')
    plt.axvline(x=mean_green, color='black', linestyle='--')
    plt.title('Histograma Verde')

    plt.subplot(2, 2, 4)
    plt.plot(hist_blue, color='blue')
    plt.axvline(x=mean_blue, color='black', linestyle='--')
    plt.title('Histograma Azul')

    plt.tight_layout()
    plt.show()
```

```python
imagen_path = "Imagen1.jpg"
Problema4_histograma(imagen_path)
```
